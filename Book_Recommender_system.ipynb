Popularity Based Recommender Systems

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace= True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg-ratings'},inplace= True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df[popular_df['num-ratings']>=250]
```

```python
popular_df = popular_df[popular_df['num-ratings']>=250].sort_values('avg-ratings',ascending = False).head(50)
```

```python
popular_df.merge(books,on="Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-S','num-ratings','avg-ratings']]
```

Collaboratice filtering

```python
c=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
good_user = c[c].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(good_user)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

Recommendation system

```python
def recommend(book_name):
  # index.fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse= True)[1:6]
  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('You Belong To Me')
```
